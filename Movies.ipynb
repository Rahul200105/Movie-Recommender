```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.shape
```

```python
movies =movies.merge(credits,on='title')
```

```python
# genres
# id
#keywords
# title
# overview
# cast
#crew
```

```python
movies =movies[['genres','id','keywords','title','overview','cast','crew']]
```

```python
movies.info()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(ob):
    l =[]
    for i in ast.literal_eval(ob):
        l.append(i['name'])
    return l
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.iloc[0].keywords
```

```python
def convert(key):
    l=[]
    for i in ast.literal_eval(key):
        l.append(i['name'])
    return(l)
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
movies.head(1)
```

```python
def first3(x):
    return x[0:3]
```

```python
movies['cast']=movies['cast'].apply(first3)
```

```python
movies.iloc[0].crew
```

```python
def Director(x):
    l=[]
    for i in ast.literal_eval(x):
        if i['job']=='Director':
            l.append(i['name'])
    return l
```

```python
movies['crew']= movies['crew'].apply(Director)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['genres'][0]
```

```python
def col(x):
    l=[]
    for i in x:
        l.append(i.replace(" ",''))
    return l
```

```python
movies['cast'] = movies['cast'].apply(col)
movies['crew'] = movies['crew'].apply(col)
movies['genres'] = movies['genres'].apply(col)
movies['keywords'] = movies['keywords'].apply(col)
```

```python
movies.head(1)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags']= movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies=movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
movies['tags']=movies['tags'].apply(lambda x: ' '.join(x))
```

```python
import nltk 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(x):
    y =[]
    for i in x.split():
        ps.stem(i)
        y.append(ps.stem(i))
    return' '.join(y)
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
(cv.get_feature_names())
```

```python
from  sklearn.metrics.pairwise import cosine_distances
```

```python
similarty =cosine_distances(vectors)
```

```python
similarty[1]
```

```python
sorted(list(enumerate(similarty[0])),key=lambda x:x[1])
```

```python
def recommend(movie):
    l=[]
    movie_index =movies[movies['title'] == movie].index[0]
    distance = similarty[movie_index]
    movie_list=sorted(list(enumerate(distance)),key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(movies.iloc[i[0]].title)
```

```python
similarty[0]
```

```python
recommend('Krrish')
```
